This notebook goes through how to **instance** geometry along the length of a **parametric curve** in 3D.

# IMPORTS

```python
import numpy as np
import matplotlib.pyplot as plt

import plotly.graph_objects as go

from spomso.cores.helper_functions import generate_grid, smarter_reshape, hard_binarization
from spomso.cores.geom_3d import Box
```

# PARAMETERS

```python
# size of the volume
co_size = 3, 3, 3
# resolution of the volume
co_resolution = 100, 100, 100

show = "FIELD" # BINARY, FIELD
show_midplane = True
show_3d = False
```

# COORDINATE SYSTEM

Point cloud of coordinates is created with the **generate_grid** function, which takes the size of the coordinate system and the number of points in each direction (**co_resolution**).

```python
coor, co_res_new = generate_grid(co_size, co_resolution)
```

If there is an even number of points along any direction the function will automatically convert it to an odd number, therefore the origin is always included in the point cloud. The point cloud is a numpy.ndarray with a shape of **(D, N)**, where **D** is the number of dimensions and **N** is the number of data points.

# CREATE SDFs

Define the **parametric function** of the **parametric curve**:

```python
def spiral(t, radius, height, freq):

    x = radius*np.cos(2*np.pi*freq*t)
    y = radius*np.sin(2*np.pi*freq*t)
    z = height*t - height/2

    return np.asarray((x, y, z))
```

**SPOMSO** has three types of instancing along a parametric curve implemented: **SIMPLE**, **ALIGNED**, **FULLY_ALIGNED**.
Set the instancing type:

```python
instancing_type = "SIMPLE"
```

Create a **box** with side lengths **0.5, 0.2, 0.3**. The **box** will be instanced on the parametric curve.

```python
box = Box(0.5, 0.2, 0.3)
```

The **box** will be instanced on the spiral **21** times.
The **radius** of the spiral is set to **1**, **height** is set to **1**, and the **frequency** to **2**:

```python
if instancing_type == "SIMPLE":
    box.curve_instancing(spiral, (1, 2, 2), (0, 1, 21))
```

It is is possible to align the **x-axis** of the **box** to the **tangent** vectors on the **curve**.

```python
if instancing_type == "ALIGNED":
    box.aligned_curve_instancing(spiral, (1, 2, 2), (0, 1, 21))
```

it is possible to align the axes of the **box** to the **tangent**, **normal** and **binormal** vectors on the curve.

```python
if instancing_type == "FULLY_ALIGNED":
    box.fully_aligned_curve_instancing(spiral, (1, 2, 2), (0, 1, 21))
```

## EVALUATE

Evaluate the SDF of the **instances** to create a signed distance field 3D map. This is done by calling the **create** method and passing in the point cloud of coordinates:

```python
spiral_pattern = box.create(coor)
```

# BINARIZATION

Convert the distance field to a binary voxel map, where 1 corresponds to the interior and 0 to the exterior of the geometry.

```python
if show_midplane:
    field = smarter_reshape(spiral_pattern, co_resolution)
    if show=="BINARY":
        pattern_2d = hard_binarization(field, 0)

if show=="BINARY":
    pattern = hard_binarization(spiral_pattern, 0)
```

# PLOT

```python
print("Drawing results...")
# Mid-plane cross-section plot
if show_midplane and show=="BINARY":
    fig, ax = plt.subplots(1,1, figsize=(8.25, 8.25))
    ax.imshow(pattern_2d[:, co_resolution[1]//2, :].T,
              cmap="binary_r",
              extent=(-co_size[0]/2, co_size[0]/2,
                      -co_size[2]/2, co_size[2]/2),
              origin="lower"
              )
    ax.grid()

    fig.tight_layout()
    plt.show()

if show_midplane and show == "FIELD":
    fig, ax = plt.subplots(1, 1, figsize=(8.25, 8.25))
    ax.imshow(field[:, co_resolution[1]//2, :].T,
              cmap="binary_r",
              extent=(-co_size[0] / 2, co_size[0] / 2,
                      -co_size[2] / 2, co_size[2] / 2),
              origin="lower"
              )
    y_mask = coor[1] == 0
    cs = ax.contour(coor[0, y_mask].reshape(co_res_new[0], co_res_new[2]),
                    coor[2, y_mask].reshape(co_res_new[0], co_res_new[2]),
                    field[:, co_resolution[1]//2, :],
                    cmap="plasma_r")
    ax.clabel(cs, inline=True, fontsize=10)
    ax.grid()

    fig.tight_layout()
    plt.show()
```

```python
# Isosurfaces plot
if show_3d and show=="BINARY":
    fig = go.Figure(data=go.Volume(
        x=coor[0],
        y=coor[1],
        z=coor[2],
        value=pattern,
        isomin=0.1,
        isomax=1,
        opacity=0.1,
        surface_count=2,
    ))
    fig.show()

if show_3d and show=="FIELD":
    fig = go.Figure(data=go.Volume(
        x=coor[0],
        y=coor[1],
        z=coor[2],
        value=spiral_pattern,
        isomin=-0,
        isomax=0.5,
        opacity=0.1,
        surface_count=5,
    ))
    fig.show()
```
